```python
#drop unrelated columns
drop_columns=['PP','Blok']
print(train_data.shape)
train_data.drop(drop_columns,axis=1,inplace=True)
print(train_data.shape)
```

```python
#drop rows where condition=R
train_data = train_data[train_data.Condition != 'R']
```

```python
#columns with missing data
missing_data_cols=list(train_data.columns[train_data.isnull().any()])
```

```python
print("num of missing data cols:", len(missing_data_cols))
```

```python
print("missing data cols:",missing_data_cols)
```

```python
#missing data
def missing_data(df):
    total_missing=df.isnull().sum().sort_values(ascending=False)
    percentage=round(df.isnull().sum().sort_values(ascending=False)*100/len(df),2)
    return pd.concat([total_missing, percentage], axis=1, keys=['Total_missing','Percentage'])
```

```python
missing_data(train_data)
```

```python
#handling missig data
for i in missing_data_cols:
    train_data[i].fillna(train_data[i].median(),inplace=True)

```

```python
missing_data(train_data)
```

```python
train_data.head(50)
```

```python
#targets
y=train_data[['Valence_rc','Arousal_rc','Dominance','Stress','MentalEffort','MentalDemand','PhysicalDemand','TemporalDemand','Effort','Performance_rc','Frustration']]
y.head()
```

```python
#drop target columns
drop_targets=['Valence_rc','Arousal_rc','Dominance','Stress','MentalEffort','MentalDemand','PhysicalDemand','TemporalDemand','Effort','Performance_rc','Frustration','NasaTLX']
print(train_data.shape)
train_data.drop(drop_targets,axis=1,inplace=True)
print(train_data.shape)
```

```python
drop_unrelated_cols=['Condition','timestamp']
print(train_data.shape)
train_data.drop(drop_unrelated_cols,axis=1,inplace=True)
print(train_data.shape)
X=train_data
X.shape
```

```python
X.head(10)
```

```python
#find categorical columns
cat_features=[i for i in X.columns if X.dtypes[i]=='object']
cat_features
```

```python
#train test split to create train set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

**Model**

**neural network**

```python

# define and fit the model
model = Sequential()
model.add(Dense(16, input_dim=156, activation='relu'))
model.add(Dropout(.2))
model.add(Dense(16, activation='relu'))
model.add(Dense(12, activation='relu'))
model.add(Dense(11, kernel_initializer='normal'))
#earlyStopping=keras.callbacks.EarlyStopping(monitor='val_loss', patience=2, verbose=0, mode='auto')
model.compile(loss='mse', optimizer='adam')
model.fit(X_train, y_train, epochs=50, verbose=0)
```

```python
#predicting the test data
prediction = model.predict(X_test)
print(prediction[1])
```

```python
#Using the rooted mean square error to evaluate the model's performance. 
score = np.sqrt(metrics.mean_squared_error(prediction, y_test))
print('Score (RMSE):{}'.format(score))
```
