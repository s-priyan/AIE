```python
# open the sample file used
file=open(r"C:\Users\sivaluxan\Desktop\FYP-2021\FYP-working\datasets\swell dataset\raw data\body postures\Kinect Studio Capture - P18_19-10-2012_c3 (2012 okt-19 11.28.30.0) - KinectData.txt","r")
  
# read the content of the file opened
content = file.readlines()
#total number of lines
number_of_lines = len(content)
print("number of lines : ",number_of_lines)
  
# read 9th line from the file
print("first line")
print(content[0])

print("second line")
print(content[1])

print("last line")
print(content[number_of_lines-1])


```

```python
header=content[0]
print(header)
#header = header.replace('\n','')
header
```

```python
#convert tab separated strig elements to list
header_list=header.split("\t")


header_list=header_list[:-1]
print("header list size:",len(header_list))
header_list
```

```python
data1=content[3]
data1
```

```python
total_data_string_list=[]
total_data_list=[]
for i in range(1,number_of_lines):
    data=content[i]
    
    #data = data.replace('\n','')
    #convert tab separated strig elements to list
    data_list=data.split("\t")
    data_list=data_list[:-1]
    total_data_string_list.append(data)
    total_data_list.append(data_list)
print("len of total data string list:",len(total_data_string_list))
print("len of total data list:",len(total_data_list))
```

```python
print("len of last element of total data list:",len(total_data_list[-1]))
total_data_list[-1]
```

```python
#create pandas data frame from list
import pandas as pd

df = pd.DataFrame (total_data_list, columns = header_list)
print (df)
```
